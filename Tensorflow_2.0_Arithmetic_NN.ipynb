```python
import pandas as pd

def separate_df(df):
  """Separates a DataFrame into two DataFrames based on data types.

  Args:
    df: The original DataFrame.

  Returns:
    A tuple of two DataFrames: (obj_df, num_df)
  """

  obj_cols = df.select_dtypes(include=['object']).columns
  num_cols = df.select_dtypes(include=['int64', 'float64']).columns

  obj_df = df[obj_cols].copy()
  num_df = df[num_cols].copy()

  return obj_df, num_df

# Example usage:

df = pd.DataFrame(df)

obj_df, num_df = separate_df(df)

print("Object DataFrame:")
print(obj_df)

print("\nNumeric DataFrame:")
print(num_df)
```

```python
# Remove outliers using IQR method
for col in num_df.columns:
    Q1 = num_df[col].quantile(0.25)
    Q3 = num_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    num_df[col] = np.where(num_df[col] < lower_bound, lower_bound,num_df[col])
    num_df[col] = np.where(num_df[col] > upper_bound, upper_bound, num_df[col])
```

```python
# Visualize the distribution and outliers after removing them
for col in num_df:
    plt.figure(figsize=(15, 4))
    plt.subplot(1, 2, 1)
    num_df[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=num_df[col])
    plt.show()
```
